- Coleção: colecao
- Biblioteca : biblioteca

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Guarnieri\Desktop\Hashtag_Python\7_days_of_code\data\Base_consolidada.csv")
display(df.head())
```

“Como se distribuem os empréstimos de exemplares pelos tipos de vínculo dos usuários?”
Desta forma, a diretoria poderá entender qual é o público que está utilizando a biblioteca e assim tomar decisões em continuar com a estratégia de negócio atual ou modificá-la.

```python
exemplares_colecao = df.groupby(['tipo_vinculo_usuario'])['id_exemplar'].count().reset_index()
exemplares_colecao.columns = ['Coleção','Quantidade de Exemplares']
exemplares_colecao = exemplares_colecao.sort_values(by='Quantidade de Exemplares', ascending=False)
display(exemplares_colecao)
```

Quais coleções são mais emprestadas?
Da mesma forma, as coleções. Ranquear as coleções mais emprestadas pelo público, será bastante importante para a estratégia atual.

```python
exemplares_colecao = df.groupby(['colecao'])['id_exemplar'].count().reset_index()
exemplares_colecao.columns = ['Coleção','Quantidade de Exemplares']
exemplares_colecao = exemplares_colecao.sort_values(by='Quantidade de Exemplares', ascending=False)
display(exemplares_colecao)
```

Quais são as bibliotecas com mais ou menos quantidade de empréstimos?
Assim, a diretoria conseguirá entender onde ela deverá melhorar e focar suas iniciativas.

```python
exemplares_colecao = df.groupby(['biblioteca'])['id_exemplar'].count().reset_index()
exemplares_colecao.columns = ['biblioteca','Quantidade de Exemplares']
exemplares_colecao = exemplares_colecao.sort_values(by='Quantidade de Exemplares', ascending=False)
display(exemplares_colecao)
```

De quais temas da CDU são os exemplares emprestados?

```python
exemplares_colecao = df.groupby(['CDU_geral'])['id_exemplar'].count().reset_index()
exemplares_colecao.columns = ['CDU_geral','Quantidade de Exemplares']
exemplares_colecao = exemplares_colecao.sort_values(by='Quantidade de Exemplares', ascending=False)
display(exemplares_colecao)
```
