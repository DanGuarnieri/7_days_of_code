```python
import pandas as pd
import os

#lista de Arquivos
caminho_pasta = r"C:\Users\Guarnieri\Desktop\Hashtag_Python\7_days_of_code\tabelas_fornecidas\dados_emprestimos"

arquivos_csv = [os.path.join(caminho_pasta, f) for f in os.listdir(caminho_pasta) if f.endswith('.csv')]

dataframes = [pd.read_csv(arquivo) for arquivo in arquivos_csv]
df_concatenado = pd.concat(dataframes, ignore_index=True)

print("Arquivos concatenados com sucesso.")
```

```python
display(df_concatenado.head())
```

```python
#Leitura do arquivo Parquet
acervo_parquet = pd.read_parquet(r'C:\Users\Guarnieri\Desktop\Hashtag_Python\7_days_of_code\tabelas_fornecidas\dados_exemplares.parquet')
display(acervo_parquet)
```

```python
#Junção dos arquivos
bases_consolidadas = df_concatenado.merge(acervo_parquet)
```

### Remoção de Duplicatas

```python
bases_consolidadas = bases_consolidadas.drop_duplicates()
```

### Removendo datas do codigo de barras

```python
bases_consolidadas['codigo_barras'] = bases_consolidadas['codigo_barras'].str.split("_").str[0]
```

### Salvar arquivo

```python
caminho_salvar = r'C:\Users\Guarnieri\Desktop\Hashtag_Python\7_days_of_code\data'
bases_consolidadas.to_csv(f"{caminho_salvar}\\Base_consolidada.csv", index=False)
```
